# Data Preparation

```python
df.drop(columns = ['panjang_total','jalan_negara'],inplace = True)
split_year = df[df.tahun.dt.year < 2023].drop(columns = ["penduduk_ribu_jiwa"])

data_backup = df.copy()
```

## Auto Regressive
Auto regressive digunakan untuk memprediksi data selain PDRB dari tahun 2023 - 2030. Model ini menggunakan lag 2 tahun sebelumnya untuk memprediksi nilai variabel tahun yang akan mendatang

```python
from statsmodels.tsa.ar_model import AutoReg
warnings.filterwarnings('ignore')

def getPredictions(df,prov,col):
  model = AutoReg(split_year[split_year.Provinsi == prov][['tahun',col]].set_index('tahun',drop = True), lags=[2,3], trend = 'c')
  model_fit = model.fit()
  return model_fit.forecast(steps = 8).reset_index().rename({'index':'tahun',0:col},axis = 1)

for prov in split_year.Provinsi.unique():
  col_res = getPredictions(split_year,prov,'persentase_sanitasi_layak')
  for col in split_year.columns.values[3:]:
    col_res = col_res.merge(getPredictions(split_year,prov,col))
  col_res['Provinsi'] = prov
  col_res['PDRB'] = np.nan
  split_year = pd.concat([split_year,col_res],axis = 0 , keys = ['Provinsi','tahun'],ignore_index=True)
```

```python
df = pd.concat([split_year,df[['penduduk_ribu_jiwa']]],axis = 1)
```

## Membersihkan Anomali

### Anomali variabel persentase
Anomali muncul ketika ada value yang diprediksi diatas 100% atau dibawah 0%.


```python
for col in ['persentase_sanitasi_layak','persentase_pengangguran','persentase_lulus_sma','persentase_air_bersih','persentase_rumah_layak','persentase penduduk_miskin']:
  print(col)
  count_beyond = len(df[df[col] > 100])
  count_under = len(df[df[col] < 0])
  print(f"(>100) : {count_beyond}")
  print(f"(<0) : {count_under}\n\n")
  if count_beyond > 0 :
    df[col] = df[col].apply(lambda x : 100 if x > 100 else x)
  if count_under < 0 :
    df[col] = df[col].apply(lambda x : 0 if x < 0  else x)
```

## Anomali varibel non - persantase
Anomali muncul ketika ada angka yang minus atau dibawah 0

```python
for col in ['jalan_provinsi', 'jalan_kabupaten_kota', 'harapan_lama_sekolah',
 'angka_harapan_hidup','pengeluaran_perkapita']:
  print(col)
  count = len(df[df[col] < 0])
  print(f"(<0) : {count}\n\n")
  if count > 0:
    df[col] = df[col].apply(lambda x : np.nan if x  < 0 else x).interpolate(method = 'cubic')
```

## Memisahkan Data Train dan Data Split

```python
train = df[df['tahun'].dt.year <= 2022]
predict = df[df['tahun'].dt.year > 2022]

train_x_index = train[['Provinsi','tahun']]

predict_x = predict.drop(columns = ['PDRB','Provinsi','tahun'])
```

## Mempersiapkan Model

```python
import xgboost as xg
from sklearn.metrics import r2_score,mean_absolute_error,mean_absolute_percentage_error
from sklearn.model_selection import StratifiedKFold,RandomizedSearchCV, GridSearchCV,train_test_split

xgb_reg = xg.XGBRegressor(objective ='reg:squarederror', n_estimators = 10,seed = 1001)
```

```python
def get_score(test_y1,test_y2):
  return pd.DataFrame({"R-squared": [r2_score(test_y1,test_y2)],
                "MAE":[mean_absolute_error(test_y1,test_y2)],
                "MAPE":[mean_absolute_percentage_error(test_y1,test_y2)]})

X_train, X_test, y_train, y_test = train_test_split(train.drop(columns = ['PDRB','Provinsi','tahun']), train['PDRB'], test_size=0.2, random_state=42)
```

```python
xgb_reg.fit(X_train,y_train)
result_1 = xgb_reg.predict(X_test)
get_score(result_1,y_test)
```

```python
def plot_average_compared(df):

  grouped_mean = df.drop(columns = "penduduk_ribu_jiwa").groupby('tahun').mean(numeric_only = True)

  names = grouped_mean.columns

  fig,axes = plt.subplots(4 , 4,figsize = (20,20))

  for i in range(len(names)):
    after_mean = df[df['tahun'].dt.year > 2022].groupby('tahun').mean(numeric_only = True)[names[i]]
    axes[i//4,i%4].set_title(f"Rata - Rata {names[i]}")
    axes[i//4,i%4].plot(grouped_mean.index,grouped_mean[names[i]])
    axes[i//4,i%4].plot(after_mean.index,after_mean)

plot_average_compared(df)
```

```python
def get_min_max(df):
  mean_province = df.groupby(['Provinsi']).mean(numeric_only=True).reset_index()
  df_min_max = pd.DataFrame({'Variable':[],'Max Value':[],'Max Province':[],'Min Value':[],'Min Province':[]})

  for col in mean_province.iloc[:,1:].columns:
      max_val = mean_province[col].max()
      min_val = mean_province[col].min()

      max_province = mean_province.loc[mean_province[col] == max_val, 'Provinsi'].values[0]
      min_province = mean_province.loc[mean_province[col] == min_val, 'Provinsi'].values[0]

      temp = pd.DataFrame([[col,max_val,max_province,min_val,min_province]],columns = ['Variable','Max Value','Max Province','Min Value','Min Province'])

      df_min_max = pd.concat([df_min_max,temp],sort = False)
  return df_min_max

get_min_max(df)
```
